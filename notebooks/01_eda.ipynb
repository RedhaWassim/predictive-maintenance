# Feature engineering 

## add rolling 

```python
train_data_processed
```

```python
window_size = 3
```

```python
taa = train_data_processed
```

```python
sensors
```

```python
# since the op settings are the settings of the engine and are almost constant ( low std ) then no need to add rolling to it
for sen in sensors:
    train_data_processed[f'{sen}_rolling_mean'] = (
        train_data_processed.groupby('unit_number')
        .rolling(window=window_size, min_periods=1)[sen]
        .mean()
        .reset_index(level=0, drop=True)
    )
    
    train_data_processed[f'{sen}_rolling_std'] = (
        train_data_processed.groupby('unit_number')
        .rolling(window=window_size, min_periods=1)[sen]
        .std()
        .reset_index(level=0, drop=True)
        .fillna(0)  # std undefined for first value
    )
```

## difference and rate of changes

For each sensor, compute the difference between the current value and the previous cycle’s value (per engine).

Why: Changes or sudden jumps in sensor readings often signal early signs of degradation.

```python
for sen in sensors : 
    train_data_processed[f'{sen}_diff'] = (
        train_data_processed.groupby('unit_number')[sen]
        .diff()
        .fillna(0) #never forget the first value can't be calculated

    )
```

# normalize cycles 

Create a normalized cycle feature: Divide time_in_cycles by the maximum cycle count for each engine.

Why: Gives a sense of how "old" the engine is relative to its full observed lifespan.

```python
train_data_processed['cycle_norm'] = (
    train_data_processed.groupby('unit_number')['time_in_cycles']
    .apply(lambda x: x / x.max())
    .reset_index(level=0, drop=True)
)
```

```python
train_data_processed
```

## Cumulative Features


Cumulative mean or sum for important sensors (shows long-term wear).

Why: Highlights overall engine wear patterns.

```python
from itertools import combinations

for a, b in combinations(sensors, 2):
    train_data_processed.loc[:, f'{a}_x_{b}'] = train_data_processed[a] * train_data_processed[b]
    train_data_processed.loc[:, f'{a}_div_{b}'] = train_data_processed[a] / (train_data_processed[b] + 1e-8)
    train_data_processed.loc[:, f'{a}_minus_{b}'] = train_data_processed[a] - train_data_processed[b]
```

```python
corr = train_data_processed.corr()
corr_columns = corr.columns
```

```python
import numpy as np

# Get upper triangle of correlation matrix
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
threshold = 0.9

# Find pairs with high correlation
high_corr_pairs = []
for col in upper.columns:
    for row in upper.index:
        if pd.notna(upper.loc[row, col]) and abs(upper.loc[row, col]) > threshold:
            high_corr_pairs.append((row, col, upper.loc[row, col]))

print("Highly correlated pairs:")
for pair in high_corr_pairs:
    print(f"{pair[0]} <-> {pair[1]}: {pair[2]:.3f}")
```

```python
# Calculate correlation of each feature with RUL
rul_corr = train_data_processed.corr()['RUL'].abs().sort_values(ascending=False)

to_drop = []
for pair in high_corr_pairs:
        col1, col2, corr_value = pair
        if col1 or col2 not in to_drop : 

            # Get correlation with RUL for both features
            col1_rul_corr = abs(rul_corr[col1])
            col2_rul_corr = abs(rul_corr[col2])
            
            # Keep the one with higher correlation to RUL, drop the other
            if col1_rul_corr > col2_rul_corr:
                to_drop.append(col2)
                print(f"Keeping {col1} (RUL corr: {col1_rul_corr:.3f}), dropping {col2} (RUL corr: {col2_rul_corr:.3f})")
            else:
                to_drop.append(col1)
                print(f"Keeping {col2} (RUL corr: {col2_rul_corr:.3f}), dropping {col1} (RUL corr: {col1_rul_corr:.3f})")

# Remove duplicates from to_drop list
to_drop = list(set(to_drop))
```
